```python
feat_col = [x for x in tr_col if x != "count"]
```

```python
X_train = train_df[feat_col]
X_train.shape
```

```python
Y_train = train_df["count"]
Y_train.shape
```

```python
rf_model = RandomForestRegressor()
```

```python
rf_model.fit(X_train, Y_train)
```

```python
predicted = rf_model.predict(X_train)
```

```python
def rmsle(orig, pred):
    ln1 = np.log(np.array(pred) + 1)
    ln2 = np.log(np.array(orig) + 1)
    return np.sqrt(np.mean((ln1 - ln2) ** 2))
```

```python
rmsle(Y_train, predicted)
```
